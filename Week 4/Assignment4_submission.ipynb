# Data source
Mortgage delinquency -  https://fred.stlouisfed.org/series/DRSFRMACBS

Unemployment - https://fred.stlouisfed.org/series/UNRATE#0

# Question
- How have the unemployment rate and mortgage delinquency prices changed over the last 26 years (1991-2017)?
- Any correlation between the two?

```python
import matplotlib.pyplot as plt
import pandas as pd

mortgage_delinquency_rate_df = pd.read_csv('DRSFRMACBS.csv') 
unemployment_rate_df = pd.read_csv('UNRATE.csv') 

# convert date column to datetime
mortgage_delinquency_rate_df['DATE'] = pd.to_datetime(mortgage_delinquency_rate_df['DATE'])
unemployment_rate_df['DATE'] = pd.to_datetime(unemployment_rate_df['DATE'])

# clip out unemployment figures older than 1991-01-01
unemployment_rate_df = unemployment_rate_df[unemployment_rate_df['DATE'] > '1990-12-31']
# clip out unemployment figures newer than 10-01-16
unemployment_rate_df = unemployment_rate_df[unemployment_rate_df['DATE'] < '2016-10-02']

# clip out delinquency figures older than 1991-01-01
mortgage_delinquency_rate_df = mortgage_delinquency_rate_df[mortgage_delinquency_rate_df['DATE'] > '1990-12-31']
# clip out delinquency figures newer than 10-01-16
mortgage_delinquency_rate_df = mortgage_delinquency_rate_df[mortgage_delinquency_rate_df['DATE'] < '2016-10-02']

# convert percentage to float
mortgage_delinquency_rate_df['DRSFRMACBS'] = pd.to_numeric(mortgage_delinquency_rate_df['DRSFRMACBS'])
unemployment_rate_df['UNRATE'] = pd.to_numeric(unemployment_rate_df['UNRATE'])

# set indices to 'DATE'
mortgage_delinquency_rate = mortgage_delinquency_rate_df.set_index(mortgage_delinquency_rate_df['DATE'])['DRSFRMACBS']
unemployment_rate = unemployment_rate_df.set_index(unemployment_rate_df['DATE'])['UNRATE']

# calculate correlation between series
#print(unemployment_rate.corr(mortgage_delinquency_rate))

# lets clip out the top and right axis
fig, ax = plt.subplots(figsize=(10,6))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(bottom='off', left='off')#, labelleft='off', labelbottom='off')


# plot unemployment rate
plt.plot(unemployment_rate, color='red', alpha=.65, label='Unemployment rate')
plt.plot(mortgage_delinquency_rate, color='blue', alpha=.65, label='Mortgage delinquency rate')
#plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=14)     # fontsize of the figure title
#plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
#plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
#plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
#plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
#plt.rc('figure', titlesize=12)  # fontsize of the figure title

# draw horizonta lines
ax.axhline(y=0, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=2, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=4, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=6, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=8, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=10, color='k', linestyle='--', lw=0.3, alpha=0.2)
ax.axhline(y=12, color='k', linestyle='--', lw=0.3, alpha=0.2)


# add some labels to the axis
#plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage', fontsize=12)

# configure the legend & title
text_str = 'Correlation = {0:.2f}'.format(unemployment_rate.corr(mortgage_delinquency_rate))
plt.legend(frameon=False)#['Unemployment rate', 'Mortgage delinquency rate'], loc=0)
plt.title('US unemployment & mortage delinquency rates between 1991 & 2017\n' + text_str)

ttl = ax.title
ttl.set_position([.5, 1.05])

#plt.figure(facecolor="white")
#plt.savefig('rate.png')
plt.show()
```
